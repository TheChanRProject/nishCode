### The Slope Intercept Form 

$$ y = mx + b $$ 

y represents the dependent variable

m represents the slope

x represents the independent variable

b represents the y-intercept 

$$ y = 5 \times x + 3 $$ 

```python
def function1(x):
    return 5*x + 3 

print(function1(1))
print(function1(2))
print(function1(3))
```

```python
import numpy as np 
```

```python
%matplotlib inline

import matplotlib.pyplot as plt 
```

```python
xvals = np.arange(-2,2, 0.01)
```

```python
yvals = function1(xvals)
```

```python
plt.plot(xvals, yvals)
plt.show()
```

Function for parabola

$$ y = x^2 $$ 

```python
def parabolic(x):
    return pow(x,2)

xvals = np.arange(-5,5,0.01)
yvals = parabolic(xvals)

plt.plot(xvals, yvals, color="red")
plt.xlabel("X Values")
plt.ylabel("Parabola Values")
plt.show()
```
